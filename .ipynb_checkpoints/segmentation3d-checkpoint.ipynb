```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os 
from scipy import ndimage as nd
from tkinter import Tcl
from poreUtils import *
```

```python
# Reading the original volume
scan_path = '1000-1300x1000-1300x2000-2300'
tiffs = os.listdir(scan_path)
slices = Tcl().call('lsort', '-dict', tiffs)
```

```python
vol = np.empty(shape=(300, 300, 300), dtype=np.uint8)
for i, fname in enumerate(slices):
    im = Image.open(os.path.join(scan_path, fname))
    imarray = np.array(im)
    imarray = np.clip(imarray, 0.0005, 0.003)
    imarray = norm8bit(imarray)
    vol[i, :, :] = imarray
```

```python
vol.shape
```

```python
plt.imshow(vol[150, :, :], cmap='gray')
```

```python
th_vol = vol < 55
th_vol = nd.binary_closing(th_vol, np.ones((3,3,3)))
th_vol = th_vol.astype(np.uint8)
```

```python
plt.imshow(th_vol[150, :, :], cmap='gray')
```

```python
import cv2
```

```python
# noise removal
opening = nd.binary_opening(th_vol, np.ones((3,3,3)))
opening = nd.binary_opening(opening, np.ones((3,3,3)))

plt.imshow(opening[150, :, :], cmap='gray')

# # sure background area
sure_bg = nd.binary_dilation(opening, np.ones((3,3,3)))
sure_bg = nd.binary_dilation(sure_bg, np.ones((3,3,3)))
sure_bg = nd.binary_dilation(sure_bg, np.ones((3,3,3)))

plt.imshow(sure_bg[150, :, :], cmap='gray')


# # Finding sure foreground area
# dist_transform = cv.distanceTransform(opening,cv.DIST_L2,5)
# ret, sure_fg = cv.threshold(dist_transform,0.7*dist_transform.max(),255,0)
# # Finding unknown region
# sure_fg = np.uint8(sure_fg)
# unknown = cv.subtract(sure_bg,sure_fg)
```

```python
img = cv2.imread('water_coins.png')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
```

```python
import skimage
```

```python
distance = nd.distance_transform_edt(thresh)
sure_fg = distance > 15
sure_fg = np.uint8(sure_fg)

markers = skimage.measure.label(sure_fg)


# coords = skimage.feature.peak_local_max(distance, footprint=np.ones((3, 3)), labels=thresh)
# mask = np.zeros(distance.shape, dtype=bool)
# mask[tuple(coords.T)] = True
# markers, _ = nd.label(mask)
labels = skimage.segmentation.watershed(-distance, markers, mask=thresh)

fig, axes = plt.subplots(ncols=3, figsize=(19, 9), sharex=True, sharey=True)
ax = axes.ravel()

ax[0].imshow(thresh, cmap=plt.cm.gray)
ax[0].set_title('Overlapping objects')
ax[1].imshow(-distance, cmap=plt.cm.gray)
ax[1].set_title('Distances')
ax[2].imshow(labels, cmap=plt.cm.nipy_spectral)
ax[2].set_title('Separated objects')

for a in ax:
    a.set_axis_off()

fig.tight_layout()
plt.show()
```
