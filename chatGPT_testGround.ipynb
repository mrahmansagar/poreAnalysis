```python
import pandas as pd
import seaborn as sns

# define three data dictionaries
data1 = {
    'Feature 1': [1, 2, 3, 4, 5],
    'Feature 2': [2, 4, 6, 8, 10],
    'Feature 3': [3, 6, 9, 12, 15],
    'Feature 4': [4, 8, 12, 16, 20]
}

data2 = {
    'Feature A': [1.0, 2.5, 3.2, 4.1, 5.8],
    'Feature B': [2.0, 4.5, 6.3, 8.1, 10.9],
    'Feature C': [3.0, 6.5, 9.4, 12.3, 15.2],
    'Feature D': [4.0, 8.5, 12.5, 16.4, 20.1]
}

data3 = {
    'Feature X': [1, 2, 3, 4, 5],
    'Feature Y': [10, 20, 30, 40, 50],
    'Feature Z': [100, 200, 300, 400, 500],
    'Feature W': [1000, 2000, 3000, 4000, 5000]
}

# concatenate the three data dictionaries into a single dataframe
df = pd.concat([pd.DataFrame(data1), pd.DataFrame(data2), pd.DataFrame(data3)], axis=1)

# calculate the correlation matrix for the combined data
corr_matrix = df.corr()

# plot the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data for multiple box plots
data = [np.random.normal(size=(100,)), np.random.normal(size=(100,)), np.random.normal(size=(100,))]

# Create the box plots
bp = plt.boxplot(data)

# Loop through the boxes and add the quartile values as text
for i, box in enumerate(bp['boxes']):
    # Calculate the quartiles
    q1, q2, q3 = np.percentile(data[i], [25, 50, 75])
    
    # Add horizontal lines for the quartiles
    # plt.axhline(q1, color='r', linestyle='--')
    # plt.axhline(q2, color='g', linestyle='-')
    # plt.axhline(q3, color='b', linestyle='--')

    # Add text labels for the quartile values
    plt.text(i+1.05, q1, f'Q1 = {q1:.2f}', color='r', fontsize=10)
    plt.text(i+1.05, q2, f'Q2 = {q2:.2f}', color='g', fontsize=10)
    plt.text(i+1.05, q3, f'Q3 = {q3:.2f}', color='b', fontsize=10)

# Show the plot
plt.show()

```
